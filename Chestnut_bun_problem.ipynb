#[Problem] The day when chestnut buns cover the solar system

##Assume that
>Diameter of a chestnut bun is 0.01 meters.

>Diameter of the solar system is 287.46 bilion kilometers



```
import math
CHESTNUT_D = 0.01
SOLAR_D = 287460000000000
def volume(d):
  return (4/3)*(d/2)**3*math.pi
```

```
chestnut_v = volume(CHESTNUT_D)
solar_v = volume(SOLAR_D)

def time_to_cover (thing_v = 1, v = 1000):
    """
    A function that find how long an input thing can cover the input volume with Byevine
    Parameteres
    --------------
    thing_v: int
        Volume of the thing (default : 1 m3)
    v: int
        Volume to covÃªr (default : 1000 m3)
        
    Returns
    --------------
    The time required in minutes
    """
    return math.ceil(math.log2(v/thing_v))*5
```

##Result

```
chestnut_time = time_to_cover(chestnut_v, solar_v)
print("The time for chestnut buns to cover the solar system: {} minutes".format(chestnut_time))
```

##Line graph

```
import matplotlib.pyplot as plt
import numpy as np
list_volume = []

def volume_per_min(v = 1, min = 100):
    """
    A function calculate the number of rice grains per minute.
    Parameteres
    --------------
    min: int
        Number of minutes to get rice (default : 100)
    Returns 
    --------------
    list_volume : list
        List of the volume of the input thing per 5 minute (the length of the list will be equal to day)
    """
    list_volume.clear()
    for i in range(min + 1):
      if i % 5 == 0:
        volume = v*2**(i//5)
        list_volume.append(volume)
    return list_volume
x = np.linspace(0, chestnut_time, chestnut_time//5 + 1, endpoint = True)
plt.figure(figsize=(20, 10))
plt.title("Chestnut buns volume per minute", fontsize = 40)
plt.xlabel("Minute", fontsize = 20)
plt.ylabel("Chestnut buns volume", fontsize = 20)
plt.tick_params(direction = 'in', length = 10, width = 1, pad = 10, labelsize = 20, colors = 'c')
plt.axhline(y = solar_v, color='r', linestyle='-')
plt.plot(x, volume_per_min(chestnut_v, chestnut_time), color = 'blue', marker = '*', linestyle = 'dotted', linewidth = 2, mec = 'black', markersize = 8)
plt.show()
```
