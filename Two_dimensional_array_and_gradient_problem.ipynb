#[Problem 1] Linear function

```python
import numpy as np
x = np.arange(-50, 50.1, 0.1)
y = 1/2 * x + 1
print("x = {}".format(x))
print("y = {}".format(y))
```

#[Problem 2] Array combination

```python
z = np.column_stack((x, y))
print("z shape = {}".format(z.shape))
print("z = {}".format(z))
```

#[Problem 3] Find the gradient

```python
grad = (y[1:] - y[:-1])/(x[1:] - x[:-1])
grad
```

#[Problem 4] Draw a graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots(2, 1, sharex = True, figsize=(20, 10))
ax[0].set_ylabel("y", fontsize = 20)
ax[0].set_title("Linear function", fontsize = 20)
ax[0].tick_params(direction = 'inout', length = 20, width = 1, pad = 10, labelsize = 15, colors = 'c')
ax[0].plot(z[:, 0], z[:, 1])
ax[0].grid()
ax[1].set_xlabel("x", fontsize = 20)
ax[1].set_ylabel("Gradient", fontsize = 20)
ax[1].set_title("Gradient", fontsize = 20)
ax[1].tick_params(direction = 'inout', length = 20, width = 1, pad = 10, labelsize = 15, colors = 'c')
ax[1].plot(z[: -1, 0], grad)
ax[1].grid()
plt.show()
```

#[Problem 5] Python functionalization

```python
def compute_gradient(function, x_range = (-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    array_xy = np.column_stack((x_range, function))
    gradient = (function[1:] - function[:-1])/(x_range[1:] - x_range[:-1])
    return array_xy, gradient

def draw_graphs(array_xy, gradient):
  fig, ax = plt.subplots(2, 1, sharex = True, figsize=(8, 6))
  ax[0].set_ylabel("y", fontsize = 20)
  ax[0].set_title("Linear function", fontsize = 20)
  ax[0].tick_params(direction = 'inout', length = 20, width = 1, pad = 10, labelsize = 15, colors = 'c')
  ax[0].plot(array_xy[:, 0], array_xy[:, 1])
  ax[0].grid()
  ax[1].set_xlabel("x", fontsize = 20)
  ax[1].set_ylabel("Gradient", fontsize = 20)
  ax[1].set_title("Gradient", fontsize = 20)
  ax[1].tick_params(direction = 'inout', length = 20, width = 1, pad = 10, labelsize = 15, colors = 'c')
  ax[1].plot(array_xy[: -1, 0], gradient)
  ax[1].grid()
  plt.show()

def function1(array_x):
  array_y = array_x**2
  return array_y

def function2(array_x):
  array_y = 2*array_x**2 + 2**array_x
  return array_y

def function3(array_x):
  array_y = np.sin(array_x**(1/2))
  return array_y

''' Compute gradient '''
x = np.arange(-50, 50.1, 0.1)
array_xy1, gradient1 = compute_gradient(function1(x), x)
array_xy2, gradient2 = compute_gradient(function2(x), x)
x = np.arange(0, 50.1, 0.1)
array_xy3, gradient3 = compute_gradient(function3(x), x)

''' Plot graphs '''
print("Graphs of function 1: \n")
draw_graphs(array_xy1, gradient1)
print("Graphs of function 2: \n")
draw_graphs(array_xy2, gradient2)
print("Graphs of function 3: \n")
draw_graphs(array_xy3, gradient3)
```

#[Problem 6] Find the minimum value

##Function 1: y = x<sup>2</sup> (50  x  50)

###Minimum value of y

```python
min1 = round(np.min(array_xy1[:, 1]), 4)
print("Minimum value of y = {} at x = {}".format(min1, np.argmin(array_xy1[:, 1])*0.1 + (-50)))
```

###Display

```python
plt.xlabel("x", fontsize = 20)
plt.ylabel("y", fontsize = 20)
plt.title("Function 1 and slope at min", fontsize = 20)
plt.tick_params(direction = 'inout', length = 20, width = 1, pad = 10, labelsize = 15, colors = 'c')
plt.plot(array_xy1[:, 0], array_xy1[:, 1])
plt.axhline(y = min1, color='r', linestyle='-')
plt.grid()
plt.show()
```

##Function 2: y = 2x<sup>2</sup> + 2<sup>x</sup> (50  x  50)

###Minimum value of y

```python
min2 = round(np.min(array_xy2[:, 1]), 4)
print("Minimum value of y = {} at x = {}".format(min2, np.argmin(array_xy2[:, 1])*0.1 + (-50)))
```

###Display

```python
plt.xlabel("x", fontsize = 20)
plt.ylabel("y", fontsize = 20)
plt.title("Function 2 and slope at min", fontsize = 20)
plt.tick_params(direction = 'inout', length = 20, width = 1, pad = 10, labelsize = 15, colors = 'c')
plt.plot(array_xy2[:, 0], array_xy2[:, 1])
plt.axhline(y = min2, color='r', linestyle='-')
plt.grid()
plt.show()
```
